```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, freqz

# Parameters
num_taps = 10  # Number of taps (filter length)
cutoff_freq = 0.2  # Cutoff frequency (normalized, where 1 corresponds to Nyquist frequency)
sampling_freq = 1.0  # Sampling frequency (normalized)

# Design the FIR filter using firwin
taps = firwin(num_taps, cutoff_freq)

# Plot the frequency response
w, h = freqz(taps, worN=8000)
freq = w / (2 * np.pi)
plt.plot(freq, 20 * np.log10(abs(h)))
plt.title('Frequency Response of FIR Filter')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Gain [dB]')
plt.grid()
plt.show()
```
